```python
import os
import csv
import numpy as np
from collections import defaultdict
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, confusion_matrix

def read_data(file_path):
    with open("/content/spamm.csv") as f:
        reader = csv.reader(f)
        next(reader) # skip header
        data = list(reader)
    return data

def preprocess_data(data):
    emails = [row[2] for row in data]
    labels = [row[1] for row in data]
    return emails, labels

def train_spam_filter(emails, labels):
    vectorizer = CountVectorizer()
    features = vectorizer.fit_transform(emails)
    train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size=0.2)

    model = MultinomialNB()
    model.fit(train_features, train_labels)

    predictions = model.predict(test_features)
    print(classification_report(test_labels, predictions))

def predict_spam(model, vectorizer, email):
    features = vectorizer.transform([email])
    prediction = model.predict(features)
    return "spam" if prediction[0] == 1 else "not spam"

if __name__ == "__main__":
    data = read_data('spam_filter_data.csv')
    emails, labels = preprocess_data(data)
    train_spam_filter(emails, labels)
```

```python
import re
import math
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('stopwords')
nltk.download('punkt')

# Initialize data
ham = []
spam = []

def preprocess(text):
    # Tokenize and remove stopwords
    words = word_tokenize(text.lower())
    words = [word for word in words if word not in stopwords.words('english')]

    # Remove punctuation
    words = [word for word in words if word.isalpha()]

    return words

def read_data(file_path):
    with open(file_path, 'r') as f:
        for line in f:
            label, email = line.split(' ', 1)
            email = preprocess(email)

            if label == 'spam':
                spam.append(email)
            else:
                ham.append(email)

def calculate_probabilities(data):
    total_emails = len(data)
    word_counts = defaultdict(int)
    total_words = 0

    for email in data:
        for word in email:
            word_counts[word] += 1
            total_words += 1

    probabilities = {}
    for word, count in word_counts.items():
        probabilities[word] = count / total_words

    return probabilities

def predict_spam(email, spam_probabilities, ham_probabilities):
    spam_score = 0
    ham_score = 0

    for word in email:
        spam_score += math.log(spam_probabilities.get(word, 0.0000001))
        ham_score += math.log(ham_probabilities.get(word, 0.0000001))

    if spam_score > ham_score:
       return 'spam'
    else:
        return 'ham'



spam_probabilities = []
ham_probabilities = []

for email, label in zip(ham, spam):
    probability = calculate_probabilities(email)

    if label == 1:
        spam_probabilities.append(probability)
    else:
        ham_probabilities.append(probability)

combined_probabilities = spam_probabilities + ham_probabilities

for probability in combined_probabilities:
    print(probability)
```
