# Câu 7  Tạo các biến TBM1, TBM2, TBM3 tương ứng với trung bình môn của các năm lớp 10, 11 và
# 12.Công thức tính: TBM = (T*2 + L + H + S + V*2 + X + D + N) / 10

```python
for i in ["1","2","6"]:
  dulieuxettuyendaihoc["TBM"+i] = (dulieuxettuyendaihoc["T"+i]*2 + dulieuxettuyendaihoc["L"+i] +dulieuxettuyendaihoc["H"+i] + dulieuxettuyendaihoc["S" + i]+ dulieuxettuyendaihoc["V"+i]*2 +dulieuxettuyendaihoc["X"+i] + dulieuxettuyendaihoc["D"+i] + dulieuxettuyendaihoc["N"+i])/10 
dulieuxettuyendaihoc.head()
```

# Câu 8 Tạo các biến xếp loại XL1, XL2 và XL3 dựa trên TBM1,TBM2 và TBM3 cho từng năm lớp
# 10, 11, 12 như sau:
#  Nhỏ hơn 5.0 xếp loại: yếu (kí hiệu là Y)
#  Từ 5.0 đến dưới 6.5: trung bình (kí hiệu là TB)
#  Từ 6.5 đến dưới 8.0: khá (kí hiệu là K)
#  Từ 8.0 đến dưới 9.0: giỏi (kí hiệu là G)
#  Từ 9.0 trở lên: xuất sắc (kí hiệu là XS)

```python
for i in ["1","2","3"] : 
  dulieuxettuyendaihoc.loc[dulieuxettuyendaihoc["TBM"+i] < 5.0,"XL"+i] = "Y" 
  dulieuxettuyendaihoc.loc[(dulieuxettuyendaihoc["TBM"+i] >= .0) & (dulieuxettuyendaihoc["TBM"+i] < 6.5),"XL"+i] ="TB"  
  dulieuxettuyendaihoc.loc[(dulieuxettuyendaihoc["TBM"+i] >=6.5) & (dulieuxettuyendaihoc["TBM"+i] < 8.0),"XL"+i] ="K" 
  dulieuxettuyendaihoc.loc[(dulieuxettuyendaihoc["TBM"+i] >= 8.0) & (dulieuxettuyendaihoc["TBM"+i] < 9.0),"XL"+i] ="G" 
  dulieuxettuyendaihoc.loc[(dulieuxettuyendaihoc["TBM"+i] >= 9.0) ,"XL"+i] ="XS"
dulieuxettuyendaihoc.head()
```

# Câu 9 Tạo các biến US_TBM1, US_TBM2 và US_TBM3 để chuyển điểm trung bình các năm lớp
# 10, 11 và 12 từ thang điểm 10 của Việt Nam sang thang điểm 4 của Mỹ.

```python
df = dulieuxettuyendaihoc
for i in df["KT"].unique():
    str(i)
    df.loc[(i == "A1" or i == "A") & (((df["DH1"]*2 + df["DH2"] + df["DH3"])/4) >= 5), "KQXT"] = 1
    df.loc[(i == "A1" or i == "A") & (((df["DH1"]*2 + df["DH2"] + df["DH3"])/4) < 5), "KQXT"] = 0

    df.loc[(i == "B") & (((df["DH1"] + df["DH2"]*2 + df["DH3"])/4) >= 5), "KQXT"] = 1
    df.loc[(i == "B") & (((df["DH1"] + df["DH2"]*2 + df["DH3"])/4) < 5), "KQXT"] = 0
    df.loc[(i not in ("A","A1","B")) & (((df["DH1"] + df["DH2"] + df["DH3"])/3) >= 5), "KQXT"] = 1
    df.loc[(i not in ("A","A1","B")) & (((df["DH1"] + df["DH2"] + df["DH3"])/3) < 5), "KQXT"] = 0
print(df.head())
```
